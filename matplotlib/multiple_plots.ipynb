## Example of multiple plots

```python
import matplotlib.pyplot as plt


x1 = [10,20,30,40,50]
x2 = [-2.0, -1.0, 0, 1.0, 2.0]

y1 = [5,10,20,30,40]
y2 = [1,3,5,7,6]

plt.subplot(2, 1, 1)
plt.plot(x1, y1)
plt.title('multiple plots')
plt.ylabel('raw values')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('normalized')

plt.show()
```
